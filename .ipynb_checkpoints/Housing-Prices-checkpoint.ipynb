# California Housing Prices

This notebook aims to solve the well known dataset of California Housing Prices from 1990 census available on Kaggle.

The dataset which we have used is the following:
https://www.kaggle.com/datasets/camnugent/california-housing-prices/data

The columns available in the dataset is:
1. longitude: A measure of how far west a house is; a higher value is farther west

2. latitude: A measure of how far north a house is; a higher value is farther north

3. housingMedianAge: Median age of a house within a block; a lower number is a newer building

4. totalRooms: Total number of rooms within a block

5. totalBedrooms: Total number of bedrooms within a block

6. population: Total number of people residing within a block

7. households: Total number of households, a group of people residing within a home unit, for a block

8. medianIncome: Median income for households within a block of houses (measured in tens of thousands of US Dollars)

9. medianHouseValue: Median house value for households within a block (measured in US Dollars)

10. oceanProximity: Location of the house w.r.t ocean/sea


The workflow which we will be following is:
1. Data Preprocessing
2. EDA
3. Training the model
4. Hyperparameter Tuning
5. Saving

## Data Preprocessing

Here we will be:
* Splitting the data into train and test sets
* Handling null values (both int and non-int) 
* Converting non-int values to int

### Split the data into train and test sets

```python
from sklearn.model_selection import train_test_split

np.random.seed(42)

x = df.drop("median_house_value", axis=1)
y = df["median_house_value"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

### Handling Null values

```python
x_train.isna().sum()
```

```python
df.isna().sum()
```

It looks like all the null values have been transferred into the test set.

So let's handle these null values directly in the test set with their data.

```python
# Fill numeric rows with median
for label, content in x_test.items():
    if pd.api.types.is_numeric_dtype(content):
        if pd.isnull(content).sum():
            x_test[label] = content.fillna(content.median())
```

```python
x_test.isna().sum()
```

```python
df.isna().sum()
```

### Converting Non-Integers into Integers

```python
x_train.dtypes
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
one_hot = OneHotEncoder()
transformer = ColumnTransformer([("one_hot", one_hot, ["ocean_proximity"])],remainder="passthrough")
```

```python
x_train_transformed = transformer.fit_transform(x_train)
x_test_transformed = transformer.fit_transform(x_test)
```

Let's create a function to cycle between all the models and give outputs rather than doing it one by one.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

models = {"Linear Regression": LinearRegression(), 
          "Decision Tree Regressor": DecisionTreeRegressor(), 
          "Random Forest Regressor": RandomForestRegressor()}

def train_model(models):
    for name, reg in models.items():
        reg.fit(x_train_transformed, y_train)
        y_pred = reg.predict(x_test_transformed)
        rmse = np.sqrt(mean_squared_error(y_test, y_pred))
        mae  = mean_absolute_error(y_test, y_pred)
        print(f"{name}")
        print(f"  RMSE: {rmse:.2f}")
        print(f"  MAE : {mae:.2f}\n")

train_model(models)
```

Before moving forward, lets create a small function called `metrics()` which will find both rmse and mae in one call rather than calling it multiple times separately.

```python
def metrics(y_true, y_pred):
    """Print RMSE and MAE given true and predicted values."""
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mae  = mean_absolute_error(y_true, y_pred)
    print(f"RMSE: {rmse:.2f}")
    print(f"MAE : {mae:.2f}")
```

## Hyperparameters Tuning

Now we will try to improve our model's performance by improving some of its hyperparameters.

We will only tune hyperparameters of `RandomForestRegressor()`

```python
from sklearn.model_selection import RandomizedSearchCV

param_dist = {
    "n_estimators": [200, 400, 600, 800, 1000],      # more trees for stability
    "max_depth": [None, 10, 20, 30, 40],             # deeper trees often help
    "min_samples_split": [2, 5, 10, 20],             # include a larger split option
    "min_samples_leaf": [1, 2, 4, 8],                # start from 1 (default) as well
    "max_features": ["sqrt", "log2", 0.5]            # numeric fraction also allowed
}


# Setup RandomizedSearchCV 
n_iter = 60 # try 60 models total
rs_clf = RandomizedSearchCV(
    estimator=model,
    param_distributions=param_dist,
    n_iter=n_iter,
    cv=5,
    verbose=2,
    n_jobs=-1,
    scoring="neg_root_mean_squared_error",
    random_state=42        # ensures reproducibility
)
rs_clf.fit(x_train_transformed, y_train)
```

```python
# Count the total number of hyperparameter combinations to test
total_randomized_hyperparameter_combintions_to_test = np.prod([len(value) for value in param_dist.values()])
print(f"There are {total_randomized_hyperparameter_combintions_to_test} potential combinations of hyperparameters to test.")
```

```python
rs_clf.best_params_
```

```python
# Make predictions with the best hyperparameters
rs_y_preds = rs_clf.predict(x_test_transformed)

# Evaluate the predictions
rs_metrics = metrics(y_test, rs_y_preds)
```
